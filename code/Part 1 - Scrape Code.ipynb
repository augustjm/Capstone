# Problem Statement

TripAdvisor aims to enhance user engagement and satisfaction by streamlining the process of finding suitable bars. Current methods require users to sift through extensive lists and reviews, which can be time-consuming and overwhelming. There is also a need to leverage the vast amount of review data to provide actionable insights to users. An intelligent chatbot can resolve these issues by offering personalized bar recommendations based on user preferences extracted from natural language input, thereby improving the decision-making process for users.

### The current problem at hand :

1. Information Overload: Users face a paradox of choice with too many options and reviews to consider.

2. Time-Consuming Process: Finding the right bar involves navigating through filters, reading numerous reviews, and comparing options manually.

3. Underutilized Data: The wealth of review data TripAdvisor possesses is not being actively used to offer personalized, data-driven recommendations.

4. User Engagement: The need to enhance user interaction with the platform, encouraging them to spend more time and return frequently.

5. Monetization Strategy: There is potential for better monetization of the platform by providing targeted advertising and featured listings.


### Benefits of the Solution:

1. Personalization: The chatbot will use natural language processing to understand user queries and preferences, offering a tailored set of recommendations.

2. Efficiency: It streamlines the decision-making process, saving users time by reducing the need to manually filter through options.

3. Data-Driven Insights: Harnessing the power of TripAdvisor's extensive review database, the chatbot can provide more accurate and relevant suggestions.

4. Increased Engagement: An interactive chatbot keeps users on the platform longer and encourages repeat usage.

5. New Revenue Streams: Featured listings and targeted advertising within the chatbot interactions can generate additional income.

6. Competitive Edge: Offering a sophisticated recommendation tool can differentiate TripAdvisor from competitors, positioning it as an innovative leader in travel tech.

7. Scalability: The chatbot can be expanded to other areas such as hotels, restaurants, and activities, making it a comprehensive travel planning tool.

```python
#Import libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
# Create a new instance of the Chrome webdriver
driver = webdriver.Chrome()

# URLS of the bar page you want to scrape
url_list = [
            "https://www.tripadvisor.com.sg/Attractions-g294265-Activities-c20-t99-oa1-Singapore.html",
            'https://www.tripadvisor.com.sg/Attractions-g294265-Activities-c20-t99-oa30-Singapore.html',
            'https://www.tripadvisor.com.sg/Attractions-g294265-Activities-c20-t99-oa60-Singapore.html',
            'https://www.tripadvisor.com.sg/Attractions-g294265-Activities-c20-t99-oa90-Singapore.html',
            'https://www.tripadvisor.com.sg/Attractions-g294265-Activities-c20-t99-oa120-Singapore.html'
           ]

# To store the bar review page links
productLinks_FrmAllPages = []

for url in url_list:

    # Open the product category page
    driver.get(url)

    # Get the current time.
    start_time = time.time()

    # Start a while loop that will run until 10 seconds have elapsed.
    while time.time() - start_time < 10:

        # Scroll the page to the bottom.
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for 1 second.
        time.sleep(1)
        
        
        
        
        
    #Extract links from each tripadvisor bar listing page
    source = driver.page_source
    data = bs(source, 'html.parser')

    # Retrieve the html code containing the product urls
    productLinks = data.find_all('a', class_='BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS')
    pos = 0
    #counter = 0

    # Retrieve the exact product urls
    for productLink in productLinks:
        # Retrieve the product url which appears every 3rd time
        #print(productLink.attrs["href"])
        if '/ShowUserReviews' in str(productLink.attrs["href"]):
            productLinks[pos] = 'https://www.tripadvisor.com.sg' + productLink.attrs["href"] + "#REVIEWS"
            print(productLinks[pos])
        else:
            productLinks[pos] = 0
        pos += 1
        #counter += 1

    # Remove all the invalid values i.e. non-review urls     
    productLinks = list(filter(lambda num: num != 0, productLinks))
    #productLinks.pop(0)

    # Print the size of data set as well as product links
    #print(len(productLinks))    
    #productLinks
    
    # Add the bar review page links for each page to the consolidated list from all pages
    productLinks_FrmAllPages.extend(productLinks)
    productLinks = []
```

```python
## Prepare a dataframe to store all the attractions and their reviews
products_df = pd.DataFrame()

# Open the attraction review page
for productLink in productLinks_FrmAllPages:
    
    review_page_link = productLink
    print(review_page_link)
    
    while len(review_page_link) != 0:
    
        driver.get(review_page_link)

        source = driver.page_source
        data = bs(source, 'html.parser')

        # Scroll the page to the bottom.
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for 1 second.
        time.sleep(1)

        # Retrieve all user review cards
        review_cards = data.find_all('div', class_='rev_wrap ui_columns is-multiline')
        review_pos = 0
        for review_card in review_cards:

            # Retrieve the values from review headings
            review_headings = data.find_all('span', class_='noQuotes')
            pos = 0
            for review_heading in review_headings:
                review_headings[pos] = review_heading.text
                pos += 1       
            #print(review_headings)    

            # Retrieve the values from review text
            review_texts = data.find_all('div', class_='prw_rup prw_reviews_text_summary_hsx')
            pos = 0
            for review_text in review_texts:
                review_texts[pos] = review_text.text
                pos += 1       
            #print(review_texts)    

            # Retrieve the values from review headings
            regex = re.compile('.*ui_bubble_rating bubble_.*')
            review_ratings = data.find_all('span', {"class" : regex})
            pos = 0
            for review_rating in review_ratings:
                review_ratings[pos] = review_rating.attrs["class"][1][-2:]#review_rating.text
                pos += 1       
            #print(review_ratings)      







        # Combine the two lists into a list of tuples
        combined_list = zip([productLink] * len(review_cards), review_headings, review_texts, review_ratings)

        # Create a DataFrame from the list of tuples
        product_df = pd.DataFrame(combined_list, columns=["Attraction", "review_heading", "review_text", "review_rating"])

        # Print the DataFrame
        #print(product_df)

        # Append the product_df to the products_df DataFrame
        products_df = products_df.append(product_df)





        # Look through the review pagination 
        review_page_link = data.find_all('div', class_='nsTKv')
        if len(review_page_link) == 0:
            if len(data.find_all('a', class_='nav next ui_button primary')) != 0:
                review_page_link = 'https://www.tripadvisor.com.sg' + data.find_all('a', class_='nav next ui_button primary')[0].attrs["href"]
            else:
                break
        print(review_page_link)   
```
